## PageRank

```python
import scipy.sparse as sp
import numpy as np
id2index = dict()
index2id = dict()
    
pure_id2friends = dict()
for uid in id2friends:
    friends = id2friends[uid]
    pure = []
    for f in friends:
        if str(f) in id2friends:
            pure.append(f)
    pure_id2friends[uid] = pure

with open("pure_id2friends.json", "w+") as fp:
    json.dump(pure_id2friends, fp)
    
for ix, uid in enumerate(pure_id2friends.keys()):
    id2index[uid] = ix
    index2id[ix] = uid

n = len(pure_id2friends)
vals = []
i_vals = []
j_vals = []
for uid in pure_id2friends:
    friends = pure_id2friends[uid]
    k = len(friends)
    if k:
        val = 1./k
    else:
        val = 0
    
    vals.extend([val]*k)
    i_vals.extend([id2index[uid]]*k)
    j_vals.extend(map(lambda i: id2index[str(i)], friends))
A = sp.coo_matrix((vals, (i_vals, j_vals)))

def power_method(A, iterations, d=.85):
    A = d*A + (1-d)/n*np.ones((n,n))

    convergence_rate = []
    x = np.ones((n,1))
    for i in range(iterations):
        x_next = A.dot(x)
        x_next /= np.linalg.norm(x_next)
        convergence_rate.append(np.linalg.norm((x_next-x))/np.linalg.norm(x))
        x = x_next
        
    # normalize pagerank to sum up to 1
#     x = x/sum(x)
    return x, convergence_rate

x, rate = power_method(A, 20)
PR_sorted_users = sorted(enumerate(x), key = lambda(_, v): v, reverse= True)
top3_PR_sorted_users = PR_sorted_users[:3]
print top3_PR_sorted_users
for i, (ix, val) in enumerate(PR_sorted_users):
    if index2id[ix] == str(5277394):
        oseledets_rank = i
        print "Oseledets: ", oseledets_rank, val
    elif index2id[ix] == str(20334973):
        caplan_rank = i
        print "Caplan: ", caplan_rank, val

with open("PR_influence.txt", "w+") as fp:
    for (ix, val) in top3_PR_sorted_users:
        fp.write(str(index2id[ix]) + ' ' + str(float(val)) + os.linesep)
    fp.write(str(5277394) + ' ' + str(0.03308169) + os.linesep)
    fp.write(str(20334973) + ' ' + str(0.03262109) + os.linesep)
```
